```python
import os

pastas_resultados = [d for d in os.listdir('../data/results') if os.path.isdir(os.path.join('../data/results', d))]
pastas_resultados.sort()

def get_environment_name(folder_path):
    """Procura um arquivo .json na pasta e retorna o nome do ambiente."""
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.json'):
            return file_name.split('-')[0]
    return "Erro?"

with open('inclusao_plots.tex', 'w') as f:
    for pasta in pastas_resultados:
        caminho_completo_pasta = os.path.join('../data/results', pasta)
        nome_ambiente = get_environment_name(caminho_completo_pasta)
        
        f.write(f'\\incluirPlotsDePasta{{{caminho_completo_pasta}}}{{{nome_ambiente}}}\n')

print('Arquivo "inclusao_plots.tex" gerado com sucesso.')
```
